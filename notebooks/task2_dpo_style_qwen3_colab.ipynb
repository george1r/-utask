# Task 2 — DPO reinforcement of the simple explanation style

This notebook first reproduces Task 1 SFT from the exact base model, then runs DPO on the
same 1,489 `kid_adult.jsonl` pairs: `kid` is `chosen` and `adult` is `rejected`. Both the
policy and its frozen reference begin from the SFT adapter trained earlier in this same run.

The held-out `public_test_style.jsonl` is not opened until SFT and DPO have both finished.
It is then used once for deterministic generation and final measurement only. There is no
system prompt or instruction asking the model to simplify.

Reproducibility contract: seed 42, greedy decoding (`do_sample=False`, `num_beams=1`),
pinned Python packages, a pinned Hugging Face model commit, explicit assertions, and no
pre-trained adapter download. CUDA/bitsandbytes training is deterministic to the extent
supported by the same Colab T4 software/hardware stack; bitwise identity across different
GPU models or CUDA kernels is not claimed.

## 5. Load and validate the training data only

The same source rows are materialized in two disjoint schemas. SFT receives only
`prompt`/`completion`, with `kid` as the assistant target. DPO receives explicit
`prompt`/`chosen`/`rejected` records, with `kid` chosen and `adult` rejected. The held-out
file is deliberately not parsed here.


```python
import hashlib

from datasets import Dataset

TRAIN_PATH = repo_root / "data" / "kid_adult.jsonl"
PUBLIC_PATH = repo_root / "data" / "public_test_style.jsonl"
STYLE_CLF_PATH = repo_root / "metrics" / "style_clf.pkl"

def read_jsonl_with_sha256(path: Path) -> tuple[list[dict], str]:
    payload = path.read_bytes()
    rows = [json.loads(line) for line in payload.decode("utf-8").splitlines() if line.strip()]
    return rows, hashlib.sha256(payload).hexdigest()

train_rows, train_sha256 = read_jsonl_with_sha256(TRAIN_PATH)
assert train_sha256 == CONFIG["kid_adult_sha256"], "Unexpected kid_adult.jsonl contents"
assert len(train_rows) == 1489, f"Expected 1489 training rows, got {len(train_rows)}"
assert all({"prompt", "kid", "adult"}.issubset(row) for row in train_rows)
for field in ("prompt", "kid", "adult"):
    assert all(isinstance(row[field], str) and row[field].strip() for row in train_rows)

sft_examples = [
    {
        "prompt": [{"role": "user", "content": row["prompt"]}],
        "completion": [{"role": "assistant", "content": row["kid"]}],
    }
    for row in train_rows
]
dpo_examples = [
    {
        "prompt": [{"role": "user", "content": row["prompt"]}],
        "chosen": [{"role": "assistant", "content": row["kid"]}],
        "rejected": [{"role": "assistant", "content": row["adult"]}],
    }
    for row in train_rows
]
sft_train_dataset = Dataset.from_list(sft_examples)
dpo_train_dataset = Dataset.from_list(dpo_examples)

assert len(sft_train_dataset) == len(dpo_train_dataset) == 1489
assert sft_train_dataset.column_names == ["prompt", "completion"]
assert dpo_train_dataset.column_names == ["prompt", "chosen", "rejected"]
assert "adult" not in sft_train_dataset.column_names
assert PUBLIC_PATH != TRAIN_PATH

print("SFT rows/columns:", len(sft_train_dataset), sft_train_dataset.column_names)
print("DPO rows/columns:", len(dpo_train_dataset), dpo_train_dataset.column_names)
print("DPO semantics: chosen=kid, rejected=adult")
print("kid_adult SHA256:", train_sha256)
print("Public test remains unopened until both trainer.train() calls finish.")
```

## 6. Load the exact Qwen model in NF4 4-bit and attach LoRA


```python
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig

tokenizer = AutoTokenizer.from_pretrained(
    CONFIG["model_id"],
    revision=CONFIG["model_revision"],
    use_fast=True,
)
assert tokenizer.chat_template, "The official Qwen chat template is required"
if tokenizer.pad_token_id is None:
    tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"

quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.float16,
)
model = AutoModelForCausalLM.from_pretrained(
    CONFIG["model_id"],
    revision=CONFIG["model_revision"],
    quantization_config=quantization_config,
    device_map={"": 0},
    torch_dtype=torch.float16,
    low_cpu_mem_usage=True,
)
assert getattr(model, "is_loaded_in_4bit", False), "Model was not loaded in 4-bit"
model.config.use_cache = False
model = prepare_model_for_kbit_training(
    model,
    use_gradient_checkpointing=True,
    gradient_checkpointing_kwargs={"use_reentrant": False},
)
model.gradient_checkpointing_enable(gradient_checkpointing_kwargs={"use_reentrant": False})

lora_config = LoraConfig(
    r=CONFIG["lora_r"],
    lora_alpha=CONFIG["lora_alpha"],
    lora_dropout=CONFIG["lora_dropout"],
    bias="none",
    task_type="CAUSAL_LM",
    target_modules=CONFIG["lora_target_modules"],
)
model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
assert set(model.peft_config["default"].target_modules) == set(CONFIG["lora_target_modules"])
```

## 8. Snapshot SFT twice: trainable policy and frozen reference

The just-trained `default` adapter is saved to temporary runtime storage, then loaded back
under the name `reference`. Tensor equality is checked before DPO. Only `default` is
trainable; the reference is not a separately downloaded model or adapter.


```python
model.save_pretrained(sft_adapter_output_dir, safe_serialization=True)
tokenizer.save_pretrained(sft_adapter_output_dir)
assert (sft_adapter_output_dir / "adapter_config.json").is_file()
assert (sft_adapter_output_dir / "adapter_model.safetensors").is_file()

model.load_adapter(
    sft_adapter_output_dir,
    adapter_name="reference",
    is_trainable=False,
)
model.set_adapter("default")
assert {"default", "reference"}.issubset(model.peft_config)

def named_adapter_parameters(adapter_name: str) -> dict[str, torch.nn.Parameter]:
    marker = f".{adapter_name}."
    return {
        name.replace(marker, ".<adapter>."): parameter
        for name, parameter in model.named_parameters()
        if marker in name
    }

policy_parameters = named_adapter_parameters("default")
reference_parameters = named_adapter_parameters("reference")
assert policy_parameters
assert policy_parameters.keys() == reference_parameters.keys()
assert all(
    torch.equal(policy_parameters[name].detach(), reference_parameters[name].detach())
    for name in policy_parameters
), "The DPO policy and reference must start from identical SFT weights"
assert all(parameter.requires_grad for parameter in policy_parameters.values())
assert not any(parameter.requires_grad for parameter in reference_parameters.values())

print("SFT adapter saved inside this runtime:", sft_adapter_output_dir)
print("Matched policy/reference adapter tensors:", len(policy_parameters))
print("Trainable policy parameters:", sum(p.numel() for p in policy_parameters.values()))
print("Frozen reference parameters:", sum(p.numel() for p in reference_parameters.values()))
print("No pretrained adapter was downloaded or loaded from an external source.")
```

```python
import gc

from trl import DPOConfig, DPOTrainer

sft_trainer.optimizer = None
sft_trainer.lr_scheduler = None
del sft_trainer
gc.collect()
torch.cuda.empty_cache()

dpo_training_output_dir = runtime_root / "task2_dpo_training"
dpo_adapter_output_dir = runtime_root / "task2_dpo_adapter"
dpo_args = DPOConfig(
    output_dir=str(dpo_training_output_dir),
    num_train_epochs=CONFIG["dpo_num_train_epochs"],
    learning_rate=CONFIG["dpo_learning_rate"],
    per_device_train_batch_size=CONFIG["per_device_train_batch_size"],
    gradient_accumulation_steps=CONFIG["gradient_accumulation_steps"],
    max_prompt_length=CONFIG["dpo_max_prompt_length"],
    max_length=CONFIG["dpo_max_length"],
    beta=CONFIG["dpo_beta"],
    loss_type=CONFIG["dpo_loss_type"],
    model_adapter_name="default",
    ref_adapter_name="reference",
    fp16=True,
    bf16=False,
    optim=CONFIG["optim"],
    warmup_ratio=CONFIG["warmup_ratio"],
    lr_scheduler_type=CONFIG["lr_scheduler_type"],
    weight_decay=CONFIG["weight_decay"],
    max_grad_norm=CONFIG["max_grad_norm"],
    logging_steps=10,
    logging_first_step=True,
    save_strategy="no",
    report_to="none",
    disable_dropout=True,
    precompute_ref_log_probs=False,
    gradient_checkpointing=True,
    gradient_checkpointing_kwargs={"use_reentrant": False},
    seed=SEED,
    data_seed=SEED,
    dataset_num_proc=1,
)
assert dpo_args.seed == 42 and dpo_args.data_seed == 42
assert dpo_args.model_adapter_name == "default"
assert dpo_args.ref_adapter_name == "reference"
assert dpo_args.loss_type == "sigmoid"

dpo_trainer = DPOTrainer(
    model=model,
    ref_model=None,
    args=dpo_args,
    train_dataset=dpo_train_dataset,
    processing_class=tokenizer,
)
assert dpo_trainer.ref_model is None
assert dpo_trainer.model_adapter_name == "default"
assert dpo_trainer.ref_adapter_name == "reference"
assert dpo_trainer.loss_type == ["sigmoid"]

dpo_train_result = dpo_trainer.train()
print("DPO training metrics:", dpo_train_result.metrics)
assert dpo_train_result.global_step > 0

reward_logs = [
    entry for entry in dpo_trainer.state.log_history
    if "rewards/margins" in entry
]
assert reward_logs, "DPO reward-margin logs were not produced"
print("First logged DPO reward margin:", reward_logs[0]["rewards/margins"])
print("Last logged DPO reward margin:", reward_logs[-1]["rewards/margins"])

model.set_adapter("default")
reference_parameters_after = named_adapter_parameters("reference")
assert not any(parameter.requires_grad for parameter in reference_parameters_after.values())
model.save_pretrained(
    dpo_adapter_output_dir,
    safe_serialization=True,
    selected_adapters=["default"],
)
assert (dpo_adapter_output_dir / "adapter_config.json").is_file()
assert (dpo_adapter_output_dir / "adapter_model.safetensors").is_file()
print("DPO policy adapter saved inside this runtime:", dpo_adapter_output_dir)
print("Public test has still not been opened.")
```

## 10. Open the held-out public test after both training stages, then generate greedily

This is the first parse of `public_test_style.jsonl`. Generation receives only the original
user `prompt`, with the official Qwen chat template and no system message or simplicity
instruction. The active adapter is the DPO-updated `default` policy.


```python
public_rows, public_sha256 = read_jsonl_with_sha256(PUBLIC_PATH)
assert public_sha256 == CONFIG["public_test_style_sha256"], "Unexpected public_test_style.jsonl contents"
assert len(public_rows) == 50, f"Expected 50 public rows, got {len(public_rows)}"
assert all({"prompt", "kid", "adult"}.issubset(row) for row in public_rows)
assert all(isinstance(row["prompt"], str) and row["prompt"].strip() for row in public_rows)
assert all(isinstance(row["kid"], str) and row["kid"].strip() for row in public_rows)
assert all(isinstance(row["adult"], str) and row["adult"].strip() for row in public_rows)
print("Held-out rows loaded after training:", len(public_rows))
print("public_test_style SHA256:", public_sha256)

import gc

dpo_trainer.optimizer = None
dpo_trainer.lr_scheduler = None
dpo_loss_types = list(dpo_trainer.loss_type)
del dpo_trainer
model.set_adapter("default")
gc.collect()
torch.cuda.empty_cache()
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)
set_seed(SEED)
model.gradient_checkpointing_disable()
model.config.use_cache = True
model.eval()

generated_answers = []
for index, row in enumerate(public_rows):
    messages = [{"role": "user", "content": row["prompt"]}]
    input_ids = tokenizer.apply_chat_template(
        messages,
        tokenize=True,
        add_generation_prompt=True,
        return_tensors="pt",
    ).to(model.device)
    attention_mask = torch.ones_like(input_ids)
    with torch.inference_mode():
        output_ids = model.generate(
            input_ids=input_ids,
            attention_mask=attention_mask,
            do_sample=False,
            num_beams=1,
            max_new_tokens=CONFIG["generation_max_new_tokens"],
            eos_token_id=tokenizer.eos_token_id,
            pad_token_id=tokenizer.pad_token_id,
            use_cache=True,
        )
    new_token_ids = output_ids[0, input_ids.shape[1]:]
    answer = tokenizer.decode(new_token_ids, skip_special_tokens=True).strip()
    generated_answers.append(answer)
    print(f"Generated {index + 1:02d}/50: {len(new_token_ids)} tokens")

assert len(generated_answers) == 50
assert all(isinstance(answer, str) and answer.strip() for answer in generated_answers)
for index in range(3):
    print(f"\nEXAMPLE {index + 1}\nPROMPT: {public_rows[index]['prompt']}\nANSWER: {generated_answers[index]}")
```

## 11. Validate `style_clf.pkl` and its feature order

The pickle is trusted task data. Its structure is checked explicitly. Reference `kid` and
`adult` texts only verify classifier direction and feature order; they never replace the
newly generated DPO answers.


```python
import joblib
from scipy.sparse import hstack

style_clf_sha256 = hashlib.sha256(STYLE_CLF_PATH.read_bytes()).hexdigest()
assert style_clf_sha256 == CONFIG["style_clf_sha256"], "Unexpected style_clf.pkl contents"
print("style_clf SHA256:", style_clf_sha256)
style_bundle = joblib.load(STYLE_CLF_PATH)
assert isinstance(style_bundle, dict)
assert {"clf", "vecs"}.issubset(style_bundle)
clf = style_bundle["clf"]
vecs = style_bundle["vecs"]
assert isinstance(vecs, (tuple, list)) and len(vecs) == 2
word_tfidf, char_tfidf = vecs
classes = np.asarray(clf.classes_)
simple_matches = np.flatnonzero(classes == 1)
assert len(simple_matches) == 1, f"Expected exactly one class labelled 1, got {classes!r}"
simple_class_index = int(simple_matches[0])

print("Classifier type:", type(clf))
print("clf.classes_:", classes)
print("Word vectorizer type:", type(word_tfidf))
print("Char vectorizer type:", type(char_tfidf))

def make_style_features(texts: list[str]):
    word_features = word_tfidf.transform(texts)
    char_features = char_tfidf.transform(texts)
    features = hstack([word_features, char_features], format="csr")
    print(
        "Feature shapes:",
        {"word": word_features.shape, "char": char_features.shape, "combined": features.shape},
    )
    assert features.shape[0] == len(texts)
    assert features.shape[1] == clf.n_features_in_
    return features

kid_features = make_style_features([row["kid"] for row in public_rows])
adult_features = make_style_features([row["adult"] for row in public_rows])
kid_sanity_mean = float(np.mean(clf.predict_proba(kid_features)[:, simple_class_index]))
adult_sanity_mean = float(np.mean(clf.predict_proba(adult_features)[:, simple_class_index]))
print(f"Sanity mean P_simple(kid): {kid_sanity_mean:.6f}")
print(f"Sanity mean P_simple(adult): {adult_sanity_mean:.6f}")
assert kid_sanity_mean > adult_sanity_mean + 0.5, "Classifier/feature-order sanity check failed"
```

## 12. Score all 50 DPO answers and show the complete table


```python
import pandas as pd
from IPython.display import display

generated_features = make_style_features(generated_answers)
per_answer_probabilities = clf.predict_proba(generated_features)[:, simple_class_index]

assert len(per_answer_probabilities) == 50
assert np.all(np.isfinite(per_answer_probabilities))
assert np.all((0.0 <= per_answer_probabilities) & (per_answer_probabilities <= 1.0))

results = pd.DataFrame(
    {
        "index": np.arange(50),
        "prompt": [row["prompt"] for row in public_rows],
        "generated_answer": generated_answers,
        "P_simple": per_answer_probabilities,
    }
)
pd.set_option("display.max_rows", 50)
pd.set_option("display.max_colwidth", None)
display(results)
assert len(results) == 50
```

## 13. Compute the Task 2 interval automatically

The four predicates follow the option order in the Task 2 statement. No metric or answer
letter is stored in advance; exactly one interval must match the arithmetic mean over all
50 generated outputs.
